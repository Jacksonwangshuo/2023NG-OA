```python
distance = data['Blocks'][2]['Geometry']['BoundingBox']['Left']-data['Blocks'][1]['Geometry']['BoundingBox']['Left']-data['Blocks'][1]['Geometry']['BoundingBox']['Width']
distance
```

```python
res = {}
for i in range(len(data['Blocks'])):
    block = data['Blocks'][i]
    if block['BlockType'] == 'LINE':
        if block['Geometry']['BoundingBox']['Left']<data['Blocks'][i-1]['Geometry']['BoundingBox']['Left']:
            res[block['Text']] = []
            tmp = block['Text']
        else:
            if len(res[tmp])>0 and block['Geometry']['BoundingBox']['Left']-data['Blocks'][i-1]['Geometry']['BoundingBox']['Left']-data['Blocks'][i-1]['Geometry']['BoundingBox']['Width']<=distance:
                res[tmp][-1]+=block['Text']
            else:
                dist = len(res[tmp])>0 and block['Geometry']['BoundingBox']['Left']-data['Blocks'][i-1]['Geometry']['BoundingBox']['Left']-data['Blocks'][i-1]['Geometry']['BoundingBox']['Width']
                num = int(dist//0.3)
                for i in range(num):
                    res[tmp].append('')
                res[tmp].append(block['Text'])
print(res)
```

```python
import pandas as pd
import json

# 读取Amazon Textract返回的JSON文件
with open('./Rule14 Interview Sample Problem/Rule14 Interview Sample Problem/OCR/testimage1.json', 'r') as f:
    data = json.load(f)

# 获取文本信息
text_list = []
for block in data['Blocks']:
    if block['BlockType'] == 'LINE':
        bbox = block['Geometry']['BoundingBox']
        polygon = block['Geometry']['Polygon']
        polygon_x = []
        polygon_y = []
        for point in polygon:
            polygon_x.append(point['X'])
            polygon_y.append(point['Y'])
        text_list.append({
            'Text': block['Text'],
            'Confidence': block['Confidence'],
            'BoundingBox_Left': bbox['Left'],
            'BoundingBox_Top': bbox['Top'],
            'BoundingBox_Width': bbox['Width'],
            'BoundingBox_Height': bbox['Height'],
            'Polygon_X': polygon_x,
            'Polygon_Y': polygon_y
        })

# 将文本信息转换为pandas DataFrame
df = pd.DataFrame(text_list)

# 输出pandas DataFrame
print(df)
```
